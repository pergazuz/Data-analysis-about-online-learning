# Function Read csv

```python
import numpy as np # นําเข้า Module numpy
import pandas as pd # นําเข้า Module pandas

Data = pd.read_csv("Database CPE102.csv",sep=",") # อ่านไฟล์ csv ที่นําเข้า
Data  # แสดงตัวอย่างข้อมูลของไฟล์ csv
```

# Function operation

```python
# สร้างตารางข้อมูลหลักโดยยังไม่ได้จําแนกด้านใด
Data_real = pd.concat([Data['เพศ'],
                       Data['ชื่อ-สกุล'],
                       Data['ที่อยู่อาศัย'],
                       Data['ได้รับผลกระทบด้านการเรียนหรือไม่'],
                       Data['ได้รับผลกระทบด้านปฏิสัมพันธ์กับคนในสังคมหรือไม่'],
                       Data['ได้รับผลกระทบด้านจิตใจและอารมณ์หรือไม่'],
                       Data['ได้รับผลกระทบด้านสุขภาพร่างกายหรือไม่'],
                       Data['ได้รับผลกระทบด้านเศรษฐกิจหรือไม่'],
                       Data['ได้รับผลกระทบด้านการทำงานในอนาคตหรือไม่'],
                       Data['ได้รับผลกระทบด้านอื่นๆ หรือไม่']],axis = 1)
                       
# สร้างตารางข้อมูลของแต่ละเพศรวมถึงหาค่าเฉลี่ยของแต่ละด้าน
Data_real_Gender = Data_real.groupby('เพศ').mean()
Data_real_Gender
```

```python
# สร้างตารางที่มีเฉพาะผลกระทบของแต่ละด้าน
Data_real2 = pd.concat([Data['ได้รับผลกระทบด้านการเรียนหรือไม่'],
                       Data['ได้รับผลกระทบด้านปฏิสัมพันธ์กับคนในสังคมหรือไม่'],
                       Data['ได้รับผลกระทบด้านจิตใจและอารมณ์หรือไม่'],
                       Data['ได้รับผลกระทบด้านสุขภาพร่างกายหรือไม่'],
                       Data['ได้รับผลกระทบด้านเศรษฐกิจหรือไม่'],
                       Data['ได้รับผลกระทบด้านการทำงานในอนาคตหรือไม่'],
                       Data['ได้รับผลกระทบด้านอื่นๆ หรือไม่']],axis = 1)
Data_real2
```

```python
# function loop แต่ละด้านเพื่อหาค่าเฉลี่ยทั้งหมดของทุกเพศ
def Data_average(Data2) :
    i = 0
    Overall = 0
    while i < 7 :
        Overall += Data2[i]
        i += 1
    Overall_average = Overall/i
    return Overall_average
```

```python
# ฟังก์ชันสร้างลิสต์ของค่าเฉลี่ยผละกระทบของแต่ละด้านจากคนที่อาศัยอยู่แถวกรุงเทพและปริมณฑล
def  Data_address_Bangkok(Data_address_Bangkok_mean) :
    # Creating an empty list
    res= []
    lis = [] 
    y = 0 
  
    # Iterating through the columns of
    # dataframe
    for column in Data_address_Bangkok_mean.columns:
      
        # Storing the rows of a column
        # into a temporary list
        li = Data_address_Bangkok_mean[column].tolist()
      
        # appending the temporary list
        res.append(li)

    z = res[0:7]
    for y in range(0,7) :
        w = z[y]
        lis.append(w[0])
    return lis
```

```python
# ฟังก์ชันสร้างลิสต์ของค่าเฉลี่ยผละกระทบของแต่ละด้านจากคนที่อาศัยอยู่แถวต่างจังหวัด
def Data_address_OutBangkok(Data_address_OutBangkok_mean) :
    # Creating an empty list
    res= []
    lis = [] 
    y = 0 
  
    # Iterating through the columns of
    # dataframe
    for column in Data_address_OutBangkok_mean.columns:
      
        # Storing the rows of a column
        # into a temporary list
        li = Data_address_OutBangkok_mean[column].tolist()
      
        # appending the temporary list
        res.append(li)

    z = res[0:7]
    for y in range(0,7) :
        w = z[y]
        lis.append(w[0])
    return lis
```

```python
# ฟังก์ชันสร้างลิสต์ของค่าเฉลี่ยผละกระทบของแต่ละด้านจากเพศชาย
def Data_Men(Data_real_Men) :
    # Creating an empty list
    res= []
    lis = [] 
    y = 0 
  
    # Iterating through the columns of
    # dataframe
    for column in Data_real_Men.columns:
      
        # Storing the rows of a column
        # into a temporary list
        li = Data_real_Men[column].tolist()
      
        # appending the temporary list
        res.append(li)

    z = res[0:7]
    for y in range(0,7) :
        w = z[y]
        lis.append(w[0])
    return  lis
```

```python
# ฟังก์ชันสร้างลิสต์ของค่าเฉลี่ยผละกระทบของแต่ละด้านจากเพศหญิง
def  Data_Women(Data_real_Women) :
    # Creating an empty list
    res = []
    lis = [] 
    y = 0 
  
    # Iterating through the columns of
    # dataframe
    for column in Data_real_Women.columns:
      
        # Storing the rows of a column
        # into a temporary list
        li = Data_real_Women[column].tolist()
      
        # appending the temporary list
        res.append(li)

    z = res[0:7]
    for y in range(0,7) :
        w = z[y]
        lis.append(w[0])
    return lis
```

```python
# ฟังก์ชันสร้างลิสต์ของค่าเฉลี่ยผละกระทบของแต่ละด้านจาก lgbtq+
def  Data_LGBTQ(Data_real_LGBTQ) :
    # Creating an empty list
    res = []
    lis = [] 
    y = 0 
  
    # Iterating through the columns of
    # dataframe
    for column in Data_real_LGBTQ.columns:
      
        # Storing the rows of a column
        # into a temporary list
        li = Data_real_LGBTQ[column].tolist()
      
    # appending the temporary list
        res.append(li)

    z = res[0:7]
    for y in range(0,7) :
        w = z[y]
        lis.append(w[0])
    return lis
```

```python
# ฟังก์ชันหาจํานวนคนที่อยู่ในกรุงเทพและต่างจังหวัด
def number_address(Data_real):
    Bangkok = Data_real[Data_real["ที่อยู่อาศัย"].isin(['กรุงเทพและปริมณฑล'])]
    Out_town = Data_real[Data_real["ที่อยู่อาศัย"].isin(['ต่างจังหวัด'])]

    return len(Bangkok.index),len(Out_town.index)
```

```python
# ฟังก์ชันหาจํานวนคนในแต่ละเพศ
def number_person(Data_real):
    men = Data_real[Data_real["เพศ"].isin(['ชาย'])]
    women = Data_real[Data_real["เพศ"].isin(['หญิง'])]
    lgbtq = Data_real[Data_real["เพศ"].isin(['LGBTQ+'])]

    return len(men.index),len(women.index),len(lgbtq.index)
```

```python
# main

# สร้างข้อมุลแยกแต่ละด้านว่ามีค่าเฉลี่ยเท่าไหร่
Data2 = Data_real.mean()

Data_address = Data_real.groupby('ที่อยู่อาศัย').mean() # สร้างตารางค่าเฉลี่ยแต่ละด้านของคนที่อยู่ในกรุงเทพ/ปริมณฑล และต่างจังหวัด
Data_address_Bangkok_mean = Data_address[0:1] # สร้าง dataframe ของคนที่อยู่ในกรุงเทพและปริมณฑล
Data_address_OutBangkok_mean = Data_address[1:2] # สร้าง dataframe ของคนที่อยู่ต่างจังหวัด

Data_real_Men = Data_real_Gender[1:2] # สร้าง dataframe ของเพศชาย
Data_real_Women= Data_real_Gender[2:3] # สร้าง dataframe ของเพศหญิง
Data_real_LGBTQ = Data_real_Gender[0:1] # สร้าง dataframe ของ LGBTQ+
```

```python
# หาค่าเฉลี่ยผู้ชาย
xm = 0
for i in range(0,7):
    xm += Data_Men(Data_real_Men)[i]
mean_men = (xm)/7
```

```python
# หาค่าเฉลี่ยผู้หญิง
xw = 0
for i in range(0,7):
    xw += Data_Women(Data_real_Women)[i]
mean_women = (xw)/7
```

```python
# หาค่าเฉลี่ย lgbtq+
xl = 0
for i in range(0,7):
    xl += Data_LGBTQ(Data_real_LGBTQ)[i]
mean_LGBTQ = (xl)/7
```

# Function Display graph and information

```python
import matplotlib.pyplot as plt

# set theme
plt.style.use("dark_background")
for param in ['text.color', 'axes.labelcolor', 'xtick.color', 'ytick.color']:
    plt.rcParams[param] = '0.9'  # very light grey
for param in ['figure.facecolor', 'axes.facecolor', 'savefig.facecolor']:
    plt.rcParams[param] = '#212946'  # bluish dark grey
colors = [
    '#08F7FE',  # teal/cyan
    '#FE53BB',  # pink
    '#FFD1DC',  # yellow
    '#00ff41',  # matrix green
]

# plot Gender graph
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
ax.axis('equal')
langs = ['MEN', 'WOMEN', 'LGBTQ+']
Gender = number_person(Data_real)
ax.pie(Gender, labels = langs,autopct='%1.2f%%' )
plt.title('Gender')
plt.show()

# plot Address graph
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
ax.axis('equal')
langs = ['Bangkok and Vicinity', 'Out of town']
address = number_address(Data_real)
ax.pie(address, labels = langs,autopct='%1.2f%%' )

plt.title('Address')
plt.show()
```

```python
#plot bars graph
bars_average = plt.bar(["ALL","MEN","WOMEN","LGBTQ+",],[Data_average(Data2),mean_men,mean_women,mean_LGBTQ] )
bars_average[0].set_color('#FFFAFA')
bars_average[1].set_color('#a2d2ff')
bars_average[2].set_color('#FFD1DC')
bars_average[3].set_color('#Cdb4db')


plt.title("Overall Average Score")
plt.xlabel("Gender")
plt.ylabel("Score")

plt.show()
print("Average all : ",Data_average(Data2))
print("Men average : ",mean_men)
print("Women average : ",mean_women)
print("LGBTQ+ average : ",mean_LGBTQ)
```

```python
Graph_address = pd.DataFrame({
   'Bangkok and Vicinity': Data_address_Bangkok(Data_address_Bangkok_mean),
   'Out of town': Data_address_OutBangkok(Data_address_OutBangkok_mean)
   }, index=["Study","Social","Passion","Health","Economy","Work","Others"])

fig, ax = plt.subplots()
Graph_address.plot(marker='o', color=colors, ax=ax)

# Redraw the data with low alpha and slighty increased linewidth:
n_shades = 10
diff_linewidth = 1.05
alpha_value = 0.3 / n_shades
for n in range(1, n_shades+1):
    Graph_address.plot(marker='o',
            linewidth=2+(diff_linewidth*n),
            alpha=alpha_value,
            legend=False,
            ax=ax,
            color=colors)
# Color the areas below the lines:
for column, color in zip(Graph_address, colors):
    ax.fill_between(x=Graph_address.index,
                    y1=Graph_address[column].values,
                    y2=[0] * len(Graph_address),
                    color=color,
                    alpha=0.1)
ax.grid(color='#2A3459') # set background color 
ax.set_xlim([ax.get_xlim()[0] - 0.2, ax.get_xlim()[1] + 0.2])  # to not have the markers cut off
ax.set_ylim(1.8)
plt.show()
```
